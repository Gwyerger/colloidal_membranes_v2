$\large\text{Generate folders, files and run batches}$

```python
import subprocess
import os
import time
```

```python
vtkpath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/VTKMega/PiWall"
energiespath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/CodeMega/DataAnalysis/PiWall/EnergyPerIteration"
snapshotpath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/CodeMega/DataAnalysis/PiWall/FinalSnapshot"
initialmeshpath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/VTKMega/PiWall"
childpath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/CodeMega/MorphoSimulations/PiWall"
parentpath = "/home/gabriel/Desktop/ProjectFiles/ColloidalMembranes/CodeMega/MorphoSimulations"
parentfilehead = "PiWallR9Main.morpho" 
```

$\text{Define parameters for the batch:}$

```python
# qchirality = [0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7]
qchirality = [0.05, 0.15, 0.25, 0.35, 0.45, 0.55, 0.65, 0.75, 0.85, 0.95, 1.05]#, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7]
```

$\text{Create new folders for morpho files and VTK exports}$

```python
if not os.path.exists(vtkpath):
    os.makedirs(vtkpath)

if not os.path.exists(energiespath):
    os.makedirs(energiespath)
    
if not os.path.exists(snapshotpath):
    os.makedirs(snapshotpath)
    
if not os.path.exists(childpath):
    os.makedirs(childpath)

for q in qchirality:
    newpath = vtkpath + f'/q={q}' 
    if not os.path.exists(newpath):
        os.makedirs(newpath)
```

$\text{Create and write to the new morpho files}$

```python
def overwrite(parentfile, childfile, q):
            with open(parentfile, 'r') as file:
                # read a list of lines into data
                data = file.readlines()

            data[17] = f"var qstar = {q} \n"
            data[28] = f'var pathexport = "{vtkpath}/q={q}"\n'
            data[29] = f'var initialmeshpath = "{initialmeshpath}"\n'
            # and write everything back
            with open(childfile, 'w') as newfile:
                newfile.writelines( data )

for q in qchirality:
    childfile = childpath + f'/PiWallR9q{q}.morpho'
    parentfile = parentpath+'/'+parentfilehead
    overwrite(parentfile, childfile, q)
```

$\text{Call the new files in ssh terminal}$

```python
print(f"Starting batch of "+parentfilehead+f" \n Batch size: {len(qchirality)} \n qs = {qchirality}")
```

```python
ti = time.time()
for q in qchirality:
    t1 = time.time()
    file = childpath + f'/PiWallR9q{q}.morpho'
    call = subprocess.call(['morpho6','-w8', file])
    t2 = time.time()
    print(f"call: {call} q = {q}, time of simulation = {(t2 - t1)/60} min \n")
tf = time.time()
print(f"time of batch: {(tf - ti)/60} min")
```
